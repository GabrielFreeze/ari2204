# Unique State-Action Counts

The x-axis shows the unique state-action. For instance, 20-10-F-H signifies:
 - Players's Hand = 20
 - Dealer's Hand = 10
 - Player Ace 11 = False
 - Action = Hit

## Random Policy

```python
random_policy_sa = pd.read_csv('random/random_sa.csv')
```

```python
random_merged = pd.DataFrame(columns=["sa","count"])
random_merged["sa"] = random_policy_sa["Player Hand"].astype(str) +'-'+ random_policy_sa["Dealer Hand"].astype(str) +'-'+ random_policy_sa["Player Ace 11"].astype(str).str[0]+'-'+random_policy_sa["Action"].astype(str).str[0]
random_merged["count"] = random_policy_sa["Times Selected"]
```

```python
random_merged.sort_values("count",ascending=False, inplace=True)
```

```python
fig = plt.figure()
fig.set_figwidth(50)
fig.set_figheight(15)
plt.bar(random_merged["sa"],random_merged["count"])
plt.xticks(rotation=90)
plt.xlabel("State-action (Player Hand, Dealer Hand, Player Ace 11 and Action)")
plt.ylabel("Count")
plt.title("Random Policy")
plt.show()
```
